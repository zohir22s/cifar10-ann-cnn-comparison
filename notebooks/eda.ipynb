# CIFAR-10 Exploratory Data Analysis (EDA)

This notebook performs exploratory data analysis on the CIFAR-10 dataset.
The goal is to understand dataset structure, class distribution, image properties,
and prepare for training the baseline ANN model.

### 1. Import libraries

```python
import torchvision
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
```

### 2. Load RAW dataset

```python
dataset = torchvision.datasets.CIFAR10(
    root="../data/raw",
    train=True,
    download=False
)

test_dataset = torchvision.datasets.CIFAR10(
    root="../data/raw",
    train=False,
    download=False
)
```

### 3. Dataset size

```python
print("Training set size:", len(dataset))
print("Test set size:", len(test_dataset))
```

The CIFAR-10 dataset contains 60,000 RGB images of size 32×32 pixels.
The dataset is divided into 50,000 training images and 10,000 test images.

### 4. Class names

```python
print("Classes:", dataset.classes)
```

```python
As we can see above, The dataset consists of 10 classes: airplane, automobile, bird, cat,
deer, dog, frog, horse, ship, and truck.
```

### 5. sample images

```python
fig, axes = plt.subplots(1, 6, figsize=(12,3))

for i in range(6):
    img, label = dataset[i]
    axes[i].imshow(img)
    axes[i].set_title(dataset.classes[label])
    axes[i].axis("off")

plt.show()
```

Figure above shows sample images from the CIFAR-10 dataset.
As observed, the dataset contains small color images representing real-world objects.

### 6. Image shape

```python
img, label = dataset[0]
print("Image size:", img.size)
print("Label:", dataset.classes[label])
```

Each image has a resolution of 32×32 pixels with three color channels (RGB).

### 7. Class distribution

```python
labels = [label for _, label in dataset]
counter = Counter(labels)

counter
```

### 8. Plot class distribution

```python
df = pd.DataFrame.from_dict(counter, orient='index')
df.index = dataset.classes

plt.figure(figsize=(10,4))
sns.barplot(x=df.index, y=df[0])
plt.title("Class Distribution")
plt.xticks(rotation=45)
plt.show()
```

The dataset seems to be balanced, with each class containing approximately 5,000 images.
This balance helps avoid bias during model training.

### 9. Pixel statistics

```python
data = np.array([np.array(img) for img, _ in dataset])

print("Mean:", data.mean())
print("Std:", data.std())
```

Pixel statistics were computed to guide normalization.
Normalization improves model convergence and stability during training.

### 10. More random samples

```python
import random

fig, axes = plt.subplots(2,5, figsize=(12,5))

for ax in axes.flatten():
    i = random.randint(0, len(dataset))
    img, label = dataset[i]
    ax.imshow(img)
    ax.set_title(dataset.classes[label])
    ax.axis("off")

plt.show()
```

## Conclusion

After the analysis we can conclude that the CIFAR-10 dataset is a balanced dataset composed of 60,000 RGB images 
divided into 10 classes. Each image has a resolution of 32×32 pixels.

The dataset is well-suited for image classification tasks. 
No significant class imbalance was observed. Pixel statistics 
justify the use of normalization before training the neural network.

This dataset will be used to train and compare between ANN,CNN and advanced CNN.
